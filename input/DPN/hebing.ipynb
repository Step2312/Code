```python
import pandas as pd
import numpy as np
```

```python
dpn = pd.read_excel('10.25-DPN患者信息表.xlsx')
```

```python
dpn.loc[0,'分组（1=DPN组，0=非DPN组）'] = '分组'
dpn.columns = dpn.loc[0]
dpn.drop(index=0,inplace=True)
dpn.reset_index(drop=True,inplace=True)
dpn
```

```python
jd0 = pd.read_excel('肌电图数据.xlsx',sheet_name='正常组')
jd1 = pd.read_excel('肌电图数据.xlsx',sheet_name='异常组')
```

```python
jd1
```

```python
jd0.iloc[3,0] = '编号'
jd0.iloc[3,1] = '姓名'
jd0.iloc[3,2] = '住院号'
jd0.columns = jd0.loc[3]
jd0.drop([0,1,2,3],inplace=True)
jd0.reset_index(drop=True,inplace=True)
jd0.loc[:,'住院号'] = jd0.loc[:,'住院号'].astype(str)
jd0.to_excel('jd0.xlsx',index=False)
```

```python
jd1.iloc[3,0] = '编号'
jd1.iloc[3,1] = '姓名'
jd1.iloc[3,2] = '住院号'
jd1.columns = jd1.loc[3]
jd1.drop([0,1,2,3],inplace=True)
jd1.reset_index(drop=True,inplace=True)
jd1.loc[:,'住院号'] = jd1.loc[:,'住院号'].astype(str)
```

```python
from collections import Counter
print(len(dpn.columns.tolist()),len(jd0.columns.tolist()),len(jd0.columns.tolist()))
col0 = dpn.columns.tolist()+jd0.columns.tolist()
col1 = dpn.columns.tolist()+jd1.columns.tolist()
col0 = [str(i) for i in col0]
col1 = [str(i) for i in col1]
f0 = open('./col0.txt',mode='w',encoding='utf-8')
f0.write(str(col0))
f0.close()
f1 = open('./col1.txt',mode='w',encoding='utf-8')
f1.write(str(col1))
f1.close()
print(str(col0)==str(col1))
```

```python
all = pd.DataFrame()

dpnqueshi = []
jdchongdu = {}
for i in dpn.loc[:,'病案号']:
    DPN_IN = False
    for j in jd0.loc[:,'住院号']:
        if str(i)==str(j):
            # print(dpn[dpn['病案号']==i].shape,jd0[jd0['住院号']==j].shape)
            if dpn[dpn['病案号']==i].shape[0] != 1:
                
                print('阴性DPN:')
                print(dpn[dpn['病案号']==i]['姓名'].tolist())
            elif jd0[jd0['住院号']==j].shape[0]!=1:
                jdchongdu.update({str(dpn[dpn['病案号']==i]['姓名'].tolist()):jd0[jd0['住院号']==j]['姓名'].tolist()})
                break
            else:
                tmp1 = dpn[dpn['病案号']==i]
                tmp1.reset_index(drop=True,inplace=True)
                tmp2 = jd0[jd0['住院号']==j]
                tmp2.reset_index(drop=True,inplace=True)
                tmp = pd.concat([tmp1,tmp2],axis=1)
                DPN_IN = True
                # print(dpn[dpn['病案号']==i].shape,jd0[jd0['住院号']==j].shape,tmp.shape)
                all = pd.concat([all,tmp])


    for j in jd1.loc[:,'住院号']:
        if str(i)==str(j):
            if dpn[dpn['病案号']==i].shape[0] != 1:
                print('阳性DPN:')
                print(dpn[dpn['病案号']==i]['姓名'].tolist())
            elif jd1[jd1['住院号']==j].shape[0]!=1:
                jdchongdu.update({str(dpn[dpn['病案号']==i]['姓名'].tolist()):jd1[jd1['住院号']==j]['姓名'].tolist()})
                break
            else:
                tmp1 = dpn[dpn['病案号']==i]
                tmp1.reset_index(drop=True,inplace=True)
                tmp2 = jd1[jd1['住院号']==j]
                tmp2.reset_index(drop=True,inplace=True)
                DPN_IN = True
                tmp = pd.concat([tmp1,tmp2],axis=1)
            all = pd.concat([all,tmp])
    if  DPN_IN is False:
        dpnqueshi = dpnqueshi+dpn[dpn['病案号']==i]['姓名'].tolist()
print(dpnqueshi,len(dpnqueshi))
print(jdchongdu,len(jdchongdu))

print(len(set(dpn.loc[:,'病案号'].tolist())),len(dpn.loc[:,'病案号'].tolist()))
print(len(set(all.loc[:,'病案号'].tolist())),len(all.loc[:,'病案号'].tolist()))
print(all.loc[:,'姓名'][all.loc[:,'病案号'].duplicated()])
all.drop_duplicates(['病案号'],inplace=True,keep=False)
print(all.shape)
all.to_excel('合并后数据.xlsx',index=False)
```
